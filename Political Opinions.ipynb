```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("political_opinions1.csv")
df2 = pd.read_csv("political_opinions2.csv")
df3 = pd.read_csv("political_opinions3.csv")
df = pd.concat([df1,df2,df3], axis = 0) #concatenation of 3 datasets
```

```python
from random import randint
Random = randint(1, 100) #Generate a random number between 1-100
df_subset = df.sample(Random) #Create Random samples from df 
df = df.drop(df_subset.index) #Drop those samples from df
df
```

```python
df = df.drop(['Timestamp'], axis = 1) #delete timestamp from dataframe
df = df.rename(columns = {"soru1":"Economy Good",
                          "soru2":"Education Reform",
                          "soru3":"Against Privatization",
                          "soru4":"Support Death Penalty",
                          "soru5":"Objective Journalists",
                          "soru6":"Alcohol Prohibition",
                          "soru7":"More Secularism",
                          "soru8":"Abortion Ban",
                          "soru9":"OHAL Restricts Freedoms",
                          "soru10":"Support New Party",
                          "Cinsiyet":"Sex",
                          "Yas":"Age",
                          "Bolge":"Region",
                          "Egitim":"Education",
                          "parti":"Party"})

df.replace(("Evet","Hayır"), ("Yes","No"), inplace = True)
df.replace(("Erkek","Kadın"), ("Male","Female"), inplace = True)
df.replace(("Marmara","Ege","Karadeniz","Akdeniz","İç Anadolu", "Doğu Anadolu", "Güneydoğu"), 
           ("Marmara","Aegean","Black Sea","Mediterrenian","Central Anatolia","Eastern Anatolia","Southeast"), inplace = True)
df.replace(("İlkokul","Ortaokul","Lise","Ön Lisans","Lisans", "Lisans Üstü"), 
           ("Primary School","Middle School","High School","Associate's Degree","University","Master's Degree"), inplace = True)
df.replace(("DIĞER"), ("OTHER"), inplace = True)
```

# DISTRIBUTION FOR THE SAMPLES 

```python
def plotGraph(attribute, title):
    attribute.value_counts().plot(kind = "pie",autopct='%.1f%%') #to show percentage of parties 
    plt.ylabel("")
    plt.title(title);
    plt.show()

plotGraph(df["Sex"], "GENDER")
plotGraph(df["Age"], "AGE")
plotGraph(df["Region"], "REGION")
plotGraph(df["Education"], "EDUCATION")
plotGraph(df["Party"], "PARTY")
```

# FINDING THE BEST CLASSIFIER

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

predictors = df.drop(["Party"], axis=1) #features
target = df["Party"] #label
x_train, x_val, y_train, y_val = train_test_split(predictors, target, test_size = 0.03, random_state= 2)
```

```python
from sklearn.tree import DecisionTreeClassifier

decisiontree = DecisionTreeClassifier(max_depth = 6)
decisiontree.fit(x_train, y_train)
y_pred = decisiontree.predict(x_val)
acc_decisiontree = round(accuracy_score(y_val, y_pred) * 100, 2)
print("Accuracy score of Decision Tree method : " , acc_decisiontree)
```

```python
from sklearn.ensemble import RandomForestClassifier

randomforest = RandomForestClassifier()
randomforest.fit(x_train, y_train)
y_pred = randomforest.predict(x_val)
acc_randomforest = round(accuracy_score(y_val, y_pred) * 100, 2)
print("Accuracy score of Random Forest method : " , acc_randomforest)
```

```python
from sklearn.svm import SVC

svm = SVC()
svm.fit(x_train, y_train)
y_pred = svm.predict(x_val)
acc_svm = round(accuracy_score(y_val, y_pred) * 100, 2)
print("Accuracy score of Support Vector method : " , acc_svm)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred = knn.predict(x_val)
acc_knn = round(accuracy_score(y_val, y_pred) * 100, 2)
print("Accuracy score of KNeighbors method : " , acc_knn)
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machine','Random Forest', 'KNeighbors', 'Decision Tree'],
    'Score': [acc_svm, acc_randomforest, acc_knn, acc_decisiontree]})

models.sort_values(by='Score', ascending=False)
```

# PLOT FOR SCORES

```python
plt.figure(figsize=(10,6))
plt.plot(models["Model"], models["Score"]);
plt.title("Model-Scores");
```
